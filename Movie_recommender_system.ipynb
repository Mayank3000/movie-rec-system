```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on="title")
```

```python
movies.head(1)
```

```python
movies = movies[['movie_id',"title",'overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast 
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
def convert_crew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew']=movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags','genres']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
same = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    dist = same[movie_index]
    movie_list = sorted(list(enumerate(dist)),reverse=True,key= lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
new_df.head()
```

```python
recommend("Batman Begins")
```

```python
genre_df=movies[['movie_id','title','genres']]
```

```python
genre_df.head()
```

```python
genre_dict = {}

# Iterate over each row in the dataframe
for index, row in genre_df.iterrows():
    for genre in row['genres']:
        if genre not in genre_dict:
            genre_dict[genre] = []
        genre_dict[genre].append(row['title'])

# Display the resulting dictionary
print(genre_dict)
```

```python
import pickle
pickle.dump(new_df,open("movies_list.pkl","wb"))
pickle.dump(same,open("same.pkl","wb"))
pickle.dump(genre_dict,open("genre_dict.pkl","wb"))
```
